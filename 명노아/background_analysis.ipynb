```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns
from matplotlib.ticker import FuncFormatter


people=pd.read_csv('./Data/인구수.csv')
people
```

```python
list(people["지 역"])
```

```python
seoul = people.loc["서울"]
cities = people.loc[["부산", "대구", "대전", "인천", "광주", "울산"]]
```

```python
for col in cities.columns:
    cities[col]=cities[col].astype(int)
seoul=seoul.astype(int)
```

```python
ratio_result=seoul/cities.sum(axis=0)
ratio_result.to_csv("서울 비율.csv")
plt.bar(cities.columns, ratio_result)
plt.xticks(range(len(cities.columns)), cities.columns, rotation=90)
plt.title("서울인구수/광역시인구수 비율 추이")
```
